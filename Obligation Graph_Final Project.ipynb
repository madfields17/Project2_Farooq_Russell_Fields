```python
import numpy as np
import pandas as pd

from pathlib import Path

test_df = pd.read_excel('fy_2022_preliminary_data_tables.xlsx', sheet_name='P-75')
test_df = test_df.apply(pd.to_numeric, errors='coerce')

import plotly.graph_objs as go
from plotly.subplots import make_subplots

#Note didn't use this in report because wanted to focus on Arrears


#Obligation Trends : P-75 

#Check for the right shape numbers
print(test_df.shape)

#Making sure I have all the years in the columns
print(test_df.columns)

#Displaying overall graph
display(test_df)

#Finding Nulls
print(test_df['2018'].isnull().sum())
print(test_df['2019'].isnull().sum())
print(test_df['2020'].isnull().sum())
print(test_df['2021'].isnull().sum())
print(test_df['2022'].isnull().sum())

#Removing Nulls- there are all on the last three columns

test_df.drop([55, 56, 57], axis = 0, inplace = True)
print(test_df)

#Find total obligation 
totals_2018 = test_df['2018'].sum()
totals_2019 = test_df['2019'].sum()
totals_2020 = test_df['2020'].sum()
totals_2021 = test_df['2021'].sum()
totals_2022 = test_df['2022'].sum()

years = ['2018', '2019', '2020', '2021', '2022']
total_five_years = [totals_2018, totals_2019, totals_2020, totals_2021, totals_2022]

#Plot the sums to show trends in the US over the last 5 years


fig = go.Figure()

fig.add_trace(go.Scatter(x = years, y = total_five_years, mode='lines+markers'))
fig.update_layout(xaxis_title = "Years", yaxis_title = "Obligation Amount (in dollars)", height=2000, showlegend=True, title_text="Trends in Collection Totals on Obligations Over Five Consecutive Years")

fig.show()

#Note to self: go back and label the x and y axis


```
